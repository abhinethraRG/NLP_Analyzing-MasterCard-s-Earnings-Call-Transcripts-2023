**Loading the NLP model for each call
transcripts. (tokenization)**


```python
%%capture
import spacy


# Load the large English NLP model for callq1
nlp1 = spacy.load("en_core_web_sm")
# Process the text using spaCy NLP
doc1 = nlp1(callq1)


# Load the large English NLP model for callq2
nlp2 = spacy.load("en_core_web_sm")
# Process the text using spaCy NLP
doc2 = nlp2(callq2)


# Load the large English NLP model for callq3
nlp3 = spacy.load("en_core_web_sm")
# Process the text using spaCy NLP
doc3 = nlp3(callq3)


# Load the large English NLP model for callq4
nlp4 = spacy.load("en_core_web_sm")
# Process the text using spaCy NLP
doc4 = nlp4(callq4)
```

**Counting the word frequency for each transcript**

Iterate over the text and count the frequency of each word (or tokens). We only process words that are not stop words, punctuation, or whitespace.

```python
%%capture

#counting the word frequency for callq1
import pandas as pd

word_frequencies1 = {}

for token in doc1:
  # Only process words that are not stop words, punctuation, or whitespace
  if not token.is_stop and not token.is_punct and not token.is_space:
      # If the word is already in the word_frequencies dictionary, increase its count
      if token.text in word_frequencies1:
          word_frequencies1[token.text] += 1
      # Otherwise, add the word to the dictionary with a count of 1
      else:
          word_frequencies1[token.text] = 1



#counting the word frequency for callq2
word_frequencies2 = {}

for token in doc2:
  # Only process words that are not stop words, punctuation, or whitespace
  if not token.is_stop and not token.is_punct and not token.is_space:
      # If the word is already in the word_frequencies dictionary, increase its count
      if token.text in word_frequencies2:
          word_frequencies2[token.text] += 1
      # Otherwise, add the word to the dictionary with a count of 1
      else:
          word_frequencies2[token.text] = 1




#counting the word frequency for callq3

word_frequencies3 = {}

for token in doc3:
  # Only process words that are not stop words, punctuation, or whitespace
  if not token.is_stop and not token.is_punct and not token.is_space:
      # If the word is already in the word_frequencies dictionary, increase its count
      if token.text in word_frequencies3:
          word_frequencies3[token.text] += 1
      # Otherwise, add the word to the dictionary with a count of 1
      else:
          word_frequencies3[token.text] = 1




#counting the word frequency for callq4

word_frequencies4 = {}

for token in doc4:
  # Only process words that are not stop words, punctuation, or whitespace
  if not token.is_stop and not token.is_punct and not token.is_space:
      # If the word is already in the word_frequencies dictionary, increase its count
      if token.text in word_frequencies4:
          word_frequencies4[token.text] += 1
      # Otherwise, add the word to the dictionary with a count of 1
      else:
          word_frequencies4[token.text] = 1
```

**Filtering and sorting for each transcript.**

Creating a dataframe from the dictionary and sort and filter the table based on frequency level 25.

Repeating this step for each transcript.

```python
#filtering and sorting -- callq1

# Use the dictionary to craete a Pandas dataframe
dfq1 = pd.DataFrame(list(word_frequencies1.items()), columns = ['Word', 'Frequency'])

#sort by frequency
dfq1 = dfq1.sort_values(by='Frequency', ascending=False)

# filter words with frequency more than equal to 25
dfq1 = dfq1.query("Frequency >= 25")

dfq1.head()
```

# **Task 3: Sentiment Analysis**

**Tokenize your transcription into a list of sentences. Here, the processed, i.e., the combined transcription is used.**

```python
#tokenization -- combined

%%capture
import nltk
nltk.download('punkt')

sentences = nltk.sent_tokenize(calls)
```

**Calculating the polarity score for each sentence using the `TextBlob` library and storing them in a list, and calculating the average polarity score.**

```python
#calculating polarity score

from textblob import TextBlob
import numpy as np

polarity_scores = []

for sentence in sentences:
    blob = TextBlob(sentence)
    polarity = blob.sentiment.polarity
    polarity_scores.append(polarity)

print("Average polarity score:", np.mean(polarity_scores))
```

**Density plot for the polarity score.**



```python
#visualizing polarity score
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(polarity_scores, shade=True)
plt.xlabel('Polarity Score')
plt.ylabel('Density')
plt.title('Density Plot of Polarity Scores')
plt.show()
```

**Polarity Summary:**
The polarity plot displays a prominent, sharp peak slightly above zero, signifying that the sentiment within the transcripts is predominantly positive, with the bulk of sentiment scores clustering around a relatively narrow positive range. This indicates that the majority of statements were optimistic, with very few negative remarks made throughout the conference calls.


**Calculating the subjectivity score for each sentence using the `TextBlob` library and storing them in a list, and calculating the average subjectivity score.**

```python
#calculating subjectivity score

from textblob import TextBlob
import numpy as np

subjectivity_scores = []

for sentence in sentences:
    blob = TextBlob(sentence)
    subjectivity = blob.sentiment.subjectivity
    subjectivity_scores.append(subjectivity)

print("Average subjectivity score:", np.mean(subjectivity_scores))
```

**Density plot for the subjectivity score.**

```python
#vizualizing subjectivity score

import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(subjectivity_scores, shade=True)
plt.xlabel('Subjectivity Score')
plt.ylabel('Density')
plt.title('Density Plot of Subjectivity Scores')
plt.show()
```

**Subjectivity Summary:**
The subjectivity plot, on the other hand, reveals a bimodal distribution with one peak near zero, indicating a substantial share of objective statements, likely reflecting factual reporting. Another peak occurs just below 0.8, suggesting that a significant portion of the discourse also included subjective remarks, potentially expressing opinions or future expectations. This dual presence of objective and subjective elements is typical in corporate communications, especially earnings calls where both data-driven results and forward-looking statements are shared.
